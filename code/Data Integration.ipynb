```python
import pandas as pd
```

```python
macro = pd.read_csv('BCI-US-UK-Dec20-Oct22.csv')
```

```python
macro
```

```python
macro_US_UK = macro[macro['TIME'] > '2020-12'][['LOCATION', 'TIME', 'Value']]
```

```python
macro_US_UK['TIME'] = macro_US_UK['TIME'].astype('datetime64[ns]')
```

```python
macro_US_UK
```

```python
snp = pd.read_csv('S&P 500 - 1Jan21 to 20Nov22.csv')
```

```python
snp
```

```python
# convert date to DATETIME to merge with macro data
snp['Date'] = snp['Date'].astype('datetime64[ns]')
```

```python
# macro and snp merged
macroAndSnp = pd.merge(snp, macro_US_UK, left_on = snp['Date'].dt.month, right_on = macro_US_UK['TIME'].dt.month)
```

```python
# unique row will be 'Date', 'LOCATION (either US or UK)'. Macro data will repeat since it is on a monthly level.
macroAndSnp
```
